### Limpieza de columnas que no vamos a necesitar

Para este caso vamos a trabajar con las siguientes columnas: Producto, Fecha, Ventas, Precio Venta, Costo Unitario Venta.

Algunos cambios que se harán son:
* Costo Unitario Venta se cambiara a Costo Unitario
* Se cambiará el nombre de los productos "Achiote Grande 250g" a "Achiote Grande", para tener consistencia de datos


```python
# Ajuste de columnas y renombre de columna Costo unitario venta
df_ventas_q1_clean = df_ventas_q1[
    ['Producto', 'Fecha', 'Ventas', 'Precio Venta', 'Costo Unitario Venta']
].rename(columns={
    'Costo Unitario Venta': 'Costo Unitario'
})

df_ventas_q1_clean.head()
```

```python
# Correción de nombre de producto Achiote Grande 250g
df_ventas_q1_clean['Producto'] = df_ventas_q1_clean['Producto'].str.strip()

df_ventas_q1_clean.loc[
    df_ventas_q1_clean['Producto'].str.contains('Achiote Grande 250', case=False, na=False),
    'Producto'
] = 'Achiote Grande'

df_ventas_q1_clean.head()
```

### Análisis y manipulación

Fase para encontrar patrones, relaciones y comportamientos



```python
# Total de ventas por producto,

ventas_por_producto = df_ventas_q1_clean.groupby('Producto')['Ventas'].sum().reset_index() #reset_index para reiniciar indice
ventas_por_producto
```

```python
# Ganancia por producto de acuerdo a las ventas por producto

# df temporal para crear columna ganancia 

df_ventas_q1_clean_temp = df_ventas_q1_clean.copy()

df_ventas_q1_clean_temp['Ganancia Total'] = (
    (df_ventas_q1_clean['Precio Venta'] - df_ventas_q1_clean['Costo Unitario']) 
    * df_ventas_q1_clean['Ventas']
)

ganancia_por_producto = df_ventas_q1_clean_temp.groupby('Producto')['Ganancia Total'].sum().reset_index()
ganancia_por_producto
```

```python
# Inversión por producto de acuerdo a las ventas por producto

# df temporal para crear columna inverisón total

df_ventas_q1_clean_temp = df_ventas_q1_clean.copy()

df_ventas_q1_clean_temp['Inversion Total'] = (
    df_ventas_q1_clean['Costo Unitario']
    * df_ventas_q1_clean['Ventas']
)

df_ventas_q1_clean_temp.head()

inversionTotal_por_producto = df_ventas_q1_clean_temp.groupby('Producto')['Inversion Total'].sum().reset_index()
inversionTotal_por_producto
```
